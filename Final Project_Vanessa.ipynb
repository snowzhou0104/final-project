## Introduction

This code will display the amount of crimes in each county to be able to determine which is the least safe and the most safe. By utlizing a bar chart, we will be able to visually see each neighboorhood is the safest and the best.

## The Metric

For this data, I will be using the column incident neighboorhood and then count the amount of times a neighboorhood is mentioned in that column. This will allow me to make bar graph to see which area in Pittsburgh has the most amount of crime in that section. So, which ever bar graph is the lowest will be marked as the best neighboorhood. This data will help determine whether the neighboorhood is safe for families.

```python
#counting neighboors
import pandas as pd
frame = str("Crime_rate.csv") 
try:
    fhand = open(frame)
except:
    print('File cannot be opened:', frame)
    exit()
    
counts = dict()

for line in fhand:
    if line.startswith('1'):
        words = line.split(',')
        count = 0
        for word in words:
            if count == 11:
                if word not in counts:
                    counts[word] = 1
                    break
                else:
                    counts[word] += 1
                    break
            count = count + 1
print(counts)
```

```python
#making the bar graph
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

chip = pd.read_csv('Crime_rate.csv', sep = ',')
chip["INCIDENTNEIGHBORHOOD"].value_counts().sort_index().plot(kind = 'bar', rot = 90, figsize = (20, 25))
```
